```python
for column in X.columns:
    plt.figure(figsize = (10,10))
    if X[column].nunique() <= 8 or X[column].dtypes == 'O':
        sns.countplot(X[column])
        plt.title(f'countplot for `{column}`')
    else:
        sns.distplot(X[column])
        plt.title(f'distribution plot for `{column}`')
    plt.show(); plt.close()
```

```python
### Feature engineering
X['Disposable_Income'] = (X['Disposable_Income'] - X['Disposable_Income'].mean()) / X['Disposable_Income'].std()

X['Total_Purchases'] = X[[c for c in X.columns if 'Purchases' in c]].aggregate('sum', axis = 1)/len([c for c in X.columns if 'Purchases' in c])
#X['Total_Purchases'] = (X['Total_Purchases'] - X['Total_Purchases'].mean()) / X['Total_Purchases'].std()

X['Total_Amount_Spent'] = X[[c for c in X.columns if 'Amount' in c]].aggregate('sum', axis = 1)/len([c for c in X.columns if 'Amount' in c])
#X['Total_Amount_Spent'] = (X['Total_Amount_Spent'] - X['Total_Amount_Spent'].mean()) / X['Total_Amount_Spent'].std()

X['Amount_Per_Purchase'] = X['Total_Amount_Spent'] / X['Total_Purchases']

X['Total_Kids'] = X['No_of_Kids_in_home'] + X['No_of_Teen_in_home']
```
